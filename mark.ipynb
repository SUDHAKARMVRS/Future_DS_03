```python
import pandas as pd

df = pd.read_csv(r"C:\Users\jaish\OneDrive\Documents\Desktop\Sudhakar\Future\bank\bank-full.csv", sep=";")

# Stage 1: Total Contacted
total_contacted = len(df)

# Stage 2: Converted
total_converted = df[df['y'] == 'yes'].shape[0]

# Drop-off
total_dropped = total_contacted - total_converted

print("Total Contacted:", total_contacted)
print("Total Converted:", total_converted)
print("Total Dropped:", total_dropped)
```

```python
import plotly.graph_objects as go

funnel = go.Figure(go.Funnel(
    y = ["Contacted", "Converted"],
    x = [total_contacted, total_converted],
    textinfo = "value+percent initial"
))

funnel.update_layout(title="Marketing Funnel Analysis")

```

```python
drop_off_rate = (total_dropped / total_contacted) * 100

print("Drop-Off Rate: {:.2f}%".format(drop_off_rate))
```

```python
contacted_more_than_once = df[df['campaign'] > 1].shape[0]

funnel_advanced = go.Figure(go.Funnel(
    y = ["Contacted", "Followed Up (Campaign >1)", "Converted"],
    x = [total_contacted, contacted_more_than_once, total_converted],
    textinfo = "value+percent previous"
))

funnel_advanced.update_layout(title="Detailed Funnel Analysis")

```
